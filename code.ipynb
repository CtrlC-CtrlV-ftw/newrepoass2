```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("water_waves_training_data.csv")

g = 9.80665

fig = plt.figure(figsize=(15, 15))

ax1 = fig.add_subplot(221, projection='3d')
ax1.set_title("3d scatter plot of height vs wavelength vs speed")
ax1.scatter(data.height, data.wave_length, data.speed)
ax1.set_xlabel('height (m)')
ax1.set_ylabel('wave length (m)')
ax1.set_zlabel('speed (m/s')

ax2 = fig.add_subplot(222, projection='3d')
ax2.set_title("3d scatter plot of dimensional homogeneous inputs")
ax2.scatter(np.sqrt(g*data.height), np.sqrt(g*data.wave_length), data.speed)
ax2.set_xlabel('sqrt( g * height )')
ax2.set_ylabel('sqrt( g * wave length )')
ax2.set_zlabel('speed')

ax3 = fig.add_subplot(223, projection='3d')
ax3.set_title("normalized 3d scatter plot of dimensional homogeneous inputs")
ax3.scatter(np.sqrt(g*data.height)*(1/data.speed), np.sqrt(g*data.wave_length)*(1/data.speed), data.speed/data.speed)
ax3.set_xlabel('sqrt( g * height ) / c')
ax3.set_ylabel('sqrt( g * wave length) / c ')
ax3.set_zlabel('speed')

ax4 = fig.add_subplot(224)
ax4.set_title("normalized scatter plot of dimensional homogeneous inputs")
ax4.scatter(np.sqrt(g*data.height)*(1/data.speed), np.sqrt(g*data.wave_length)*(1/data.speed))
ax4.set_xlabel('sqrt( g * height ) / c')
ax4.set_ylabel('sqrt( g * wave length ) / c')

plt.tight_layout()
plt.show()
```
